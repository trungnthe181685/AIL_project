```python
import pandas as pd
import matplotlib.pyplot as plt
# Đọc dữ liệu
data = pd.read_csv("ratings.csv")
```

```python
# 1. Kiểm tra thông tin tổng quan
print("Thông tin tổng quan:")
data.info()
```

```python
# 2. Mô tả thống kê
print("\nMô tả thống kê:")
print(data.describe())
```

```python
# 3. Kiểm tra giá trị thiếu
print("\nSố lượng giá trị thiếu:")
print(data.isnull().sum())
```

```python
# 4. Phân phối của cột rating
print("\nPhân phối điểm đánh giá:")
print(data["rating"].value_counts().sort_index())
```

```python
rating_distribution = data['rating'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
plt.bar(rating_distribution.index, rating_distribution.values, color='skyblue')
plt.xticks(rating_distribution.index)  # Hiển thị các nhãn rating rõ ràng
plt.xlabel('Rating')
plt.ylabel('Số lượng')
plt.title('Phân bố của đánh giá')
plt.show()
```

```python
# 5. Tần suất đánh giá theo sản phẩm (Top 10)
print("\nTop 10 sản phẩm được đánh giá nhiều nhất:")
item_frequency = data["item"].value_counts().head(10)
print(item_frequency)
```

```python
plt.figure(figsize=(10, 5))
item_frequency.head(10).plot(kind='bar')
plt.xlabel('Item')
plt.ylabel('Số lượt đánh giá')
plt.title('Top 10 sản phẩm/khóa học được đánh giá nhiều nhất')
plt.show()
```

```python
# 6. Số lượng đánh giá của từng người dùng (Top 10)
print("\nTop 10 người dùng có nhiều đánh giá nhất:")
user_activity = data["user"].value_counts().head(10)
print(user_activity)
```

```python
plt.figure(figsize=(10, 5))
user_activity.head(10).plot(kind='bar', color='orange')
plt.xlabel('User')
plt.ylabel('Số lượt đánh giá')
plt.title('Top 10 người dùng có nhiều đánh giá nhất')
plt.show()
```
